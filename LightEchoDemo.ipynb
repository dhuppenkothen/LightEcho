## Demos for the LightEcho code

This notebook describes the echo state network in LightEcho and how to use it. We'll reproduce results from Giannotis+ 2015, found here: http://arxiv.org/abs/1505.00936
Code in here requires:
- numpy
- scipy
- scikit-learn
- matplotlib

The class tries to reproduce the scikit-learn API as well as I know how to, but no guarantees that it's actually true (yet). 


```python
## Let's do some imports first
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import numpy as np
import sys
sys.path.append("code/")

from echostate import EchoStateNetwork

from sklearn.grid_search import GridSearchCV
```

The Echo State Network has several hyperparameters that need to be set when instantiating the class:
- `N` sets the number of hidden units in the network
- `a` sets the absolute value for the input weights (the sign will be set by a deterministic, aperiodic sequence generated by a random number generator with a fixed seed)
- `r` sets the value of the forward connections between hidden units; the weight will be the same for all connections that exist, and zero otherwise
- `n_washout`: the number of samples at the beginning of the time series to dismiss when fitting for the output weights
- `scaling`: sets the memory between consecutive time steps (the value of the hidden units in a current time step will be the activation of the units at that time step plus `scaling` times the value of the units in the previous time step
- `lamb`: regularisation parameter for ridge regression
- `seed`: the seed for the random number generator to generate the deterministic, aperiodic sequence for generating input and hidden weights
- `b` sets the weight for the backwards connections of hidden units in any topology that requires these backwards connections
- `topology` sets the topology of the hidden reservoir; `scr` for simple circle reservoir (units organised in a circle), `dlr` for delay line reservoir (units organised in a line) or `dlrb` for delay line reservoir with backward connections (same as DLR, but with backward connections)

In particular `r`, `a` and `N` should be set by cross validation or any other method to find hyperparameters.

### Setting up simulated NARMA data

Following Giannotis+ 2015, we'll do some tests on simulated data and show how the class works in the process.
We'll first generate data from a NARMA (Non-linear Auto-Regressive Moving Average) process, where the current output depends both on the input and the previous output. In practice, systems of this type are hard to model because of their non-linearity and potential long-term memory.

Let's define the NARMA classes:

```python
def narma1(time):
    ## randomly generated inputs
    s = np.random.uniform(0.0,0.5, size=time.shape[0])
    
    ## set NARMA class 
    narma_class = 9 
    
    ## set up array for the outputs
    y = np.zeros(time.shape[0])
    ## loop through all elements
    for t in xrange(0,y.shape[0]-1):
        firstterm = 0.3*y[t]
        sumterm = np.sum([y[t-i] for i in xrange(narma_class + 1)])
        secondterm = 0.05*y[t]*sumterm
        thirdterm = 1.5*s[t-narma_class]*s[t]
        y[t+1] = firstterm + secondterm + thirdterm + 0.1
    return y

def narma2(time):
    s = np.random.uniform(0.0, 0.5, size=time.shape[0])
    y = np.zeros(time.shape[0])
    print(y.shape)
    narma_class = 19
    for t in xrange(0,y.shape[0]-1):
        firstterm = 0.3*y[t]
        sumterm = np.sum([y[t-i] for i in xrange(narma_class+1)])
        second_term = 0.05*y[t]*sumterm
        thirdterm = 1.5*s[t-narma_class]*s[t]
        tanhterm = np.tanh(firstterm + second_term + thirdterm + 0.01)
        y[t+1] = tanhterm + 0.2
    return y

def narma3(time):
    s = np.random.uniform(0.0, 0.5, size=time.shape[0])
    y = np.zeros(time.shape[0])
    narma_class = 29
    for t in xrange(0,y.shape[0]-1):
        firstterm  = 0.2*y[t]
        sumterm = np.sum([y[t-i] for i in xrange(narma_class+1)])
        secondterm = 0.004*s[t]*sumterm
        thirdterm = 1.5*s[t-narma_class]*s[t] + 0.201
        y[t+1] = firstterm + secondterm + thirdterm
    return y
```

```python
times = np.arange(800)
y1 = narma1(times)
s = np.random.uniform(0.0, 0.5, size=times.shape[0])
plt.figure()
plt.plot(times, y1)

y2 = narma2(times)
plt.figure()
plt.plot(times, y2)

y3 = narma3(times)
plt.figure()
plt.plot(times, y3)

```

### A simple Echo State Network

We'll take the first of the three NARMA processes and generate a light curve to model:

```python
times = np.arange(1000)
y = narma1(times)
```

Now we can set up the Echo State Network. We'll set the hyperparameters to some values for now; we can test them later.
They all have default parameters, so theoretically you can leave them be. However, no guarantee that the default values are useful for your particular problem, so you're probably better off using whatever knowledge you might have, or use cross-validation to find the best-use values for your problem.

```python
N = 200 ## number of hidden units
a = 0.5 ## input unit weights
r = 0.1 ## hidden unit weights
n_washout = 100 ## number of washout samples
scaling=1.0 ## scaling between current and previous iteration; 1.0 means no scaling
lamb = 1.e-4 ## regularization parameter
seed = 20150519 ## seed for random number generator
b = None ## weights of backward connections between hidden units; not used in SCR topology
topology = "scr" ## use Simple Circle Reservoir topology

esn = EchoStateNetwork(N, a, r, n_washout, scaling, lamb, seed, b, topology)

```

We can now fit some data set. Here, we use one-dimensional time series, as is typical for astronomy, but we could use multi-dimensional data. If you use the latter, please make sure that your data is in a numpy-array of forma (Nsamples, Ndimensions), otherwise the code will break!

```python
esn.fit(y)
ww = esn.transform(y)
```

Let's see how well the network predicts a different time series of the same type:

```python
ytest = narma1(times)
ypred = esn.predict(ytest)
plt.plot(ytest, color="black", label="data")
plt.plot(ypred, color="red", label="ESN prediction")
```

Not bad! Can we quantify this? Yes, we can!
We'll use the normalized mean square error (NSME), implemented in scikit-learn as `r2_score` and in the `EchoStateNetwork` class as the `score`-method.

```python
print("Prediction NSME score: %.3f"%esn.score(ytest))
```

Since the best possible score is 1.0, this isn't super-great, even though the plot looks quite good. 
We can make this slightly better by scaling the train and test data sets to a mean of zero and a variance of one.

```python
from sklearn.preprocessing import StandardScaler
yall = np.hstack((y, ytest))
yall_scaled = StandardScaler().fit_transform(yall)
yscaled = yall_scaled[:1000]
ytest_scaled = yall_scaled[1000:]
esn.fit_transform(yscaled)
print("Prediction NSME score: %.3f"%esn.score(ytest_scaled))

yscaled_pred = esn.predict(ytest_scaled)
plt.plot(ytest_scaled, color="black", label="data")
plt.plot(yscaled_pred, color="red", label="ESN prediction")
```

### Cross-Validation

We'd like to choose better hyperparameters than the random guesses we've made above in order to improve our predictions.
For this, we'll use `GridSearchCV` form scikit-learn to help us.


```python
y = narma1(np.arange(1000))
import echostate
reload(echostate)
tuned_parameters = {"N":[10,50,100,200], "a":[0.1, 0.5,  1.0, 2.0,], 
                    "r": [0.01, 0.1, 0.5, 1.0],
                    "scaling":[0.1,0.5,0.9], "lamb":[0.0001, 0.001, 0.01, 0.1,]}

clf = GridSearchCV(echostate.EchoStateNetwork(n_washout=20), param_grid = tuned_parameters, cv=2)
clf.fit(y)
```

```python
print("The best parameters are: " + str(clf.best_params_))
print("The score for the best parameter ESN is: " + str(clf.best_score_))
```

```python
esn_best = clf.best_estimator_
```
